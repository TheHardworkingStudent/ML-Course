## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize= (10,10)) # (width, height)

# 3. Plot data
ax.plot(x,y)
# 4. Customize plot
ax.set(title = "Simple Plot", xlabel = "x-axis", ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows = 2, ncols = 3, figsize = (15, 7))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
ax5.set(title = "empty set", xlabel = "no data");
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (15, 7))
ax[0][0].plot(x, x/2);
ax[0][1].scatter(np.random.random(10), np.random.random(10));
ax[0][2].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1][0].hist(np.random.randn(1000));
ax[1][1].set(title = "empty set", xlabel = "no data");
```

## Which one should you use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use the pyplot method
* When plotting someth ing more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target');
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);
#ax.set_xlim([45,100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"]);

# Customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")
```
