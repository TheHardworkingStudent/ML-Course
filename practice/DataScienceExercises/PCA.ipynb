```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Original Dataset")
plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(iris[['sepal_length', 'sepal_width']])

# Perform PCA with two components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Create a DataFrame with the PCA results
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
pca_df['species'] = iris['species']

sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="species")
plt.xlabel("Principal Component 1 (PC1)")
plt.ylabel("Principal Component 2 (PC2)")
plt.title("PCA Transformed Data")
plt.show()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
component_labels = ['PC1', 'PC2']

sns.barplot(x=component_labels, y=explained_variance_ratio)
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance Ratio")
plt.title("Variance Explained by Principal Components")
plt.ylim(0, 1)
plt.show()
```
